15. Даны x = [1,2], y = [[4, 1], [2, 2]]. Найдите произведение матриц x * y и y * x.

```python
arr = np.array([1,2])
second_arr = np.array([[4,1], [2,2]])
print(np.dot(arr, second_arr))
print(np.dot(second_arr, arr))
```

16. Даны x = [[1, 0], [0, 1]], y = [[4, 1], [2, 2], [1, 1]]. Найдите произведение матриц.

```python
arr = np.array([[1,0], [0,1]])
second_arr = np.array([[4,1], [2,2], [1,1]])
print(np.dot(arr, second_arr))
```

17. Создайте массив x, состоящий из 10 чисел от 0 до 9. Если элемент x меньше 3, замените его на 3. И если элемент x больше 7, замените его на 7. 

```python
arr = np.arange(0,10,1)
print(arr)
arr[arr<3] = 3
arr[arr>7] = 7
print(arr)
```
